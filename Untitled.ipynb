```python
import serial as ser
import numpy as np
import laspy
import serial.tools.list_ports
ports = serial.tools.list_ports.comports()

for port, desc, hwid in sorted(ports):
        print("{}: {} [{}]".format(port, desc, hwid))
```

```python
strdata = []
# arduino = ser.Serial('/dev/ttyUSB1', baudrate = 9600, timeout=.1) #assuming serial connection is (COM3)
curr = 0
counter = 0
x, y, z = 0, 0, 0
allX = []
allY = []
allZ = []
S = 'S'
loop = True
```

```python
arduino = ser.Serial('/dev/ttyUSB2', baudrate = 9600, timeout=.1) #assuming serial connection is (COM3)
arduino.write(S.encode('utf-8'))
while loop:
    temp = arduino.readline().decode('ascii')
    print(temp)
    strdata.append(temp)
    if temp == "Stop":
        loop = False
print(strdata)
```

```python
strdata = [x.replace("\r\n","") for x in strdata]
strdata = [ x for x in strdata if "," in x ]
data = [ x.split(",") for x in strdata]
data = np.array(data).astype(float)
data
```

```python
data1 = np.zeros(data.shape)
data1[:,0] = data[:,0]*np.cos(data[:,2]*np.pi/180)*np.cos(data[:,1]*np.pi/180)
data1[:,1] = data[:,0]*np.cos(data[:,2]*np.pi/180)*np.sin(data[:,1]*np.pi/180)
data1[:,2] = data[:,0]*np.sin(data[:,2]*np.pi/180)
```

```python
hdr = laspy.header.Header()

lasOutput = laspy.file.File("lidarang.las", mode = 'w', header = hdr)


xMin = np.floor(np.min(data1[:,0])) #use numpy to get int min values of x y z
yMin = np.floor(np.min(data1[:,1]))
zMin = np.floor(np.min(data1[:,2]))

lasOutput.header.offset = [xMin, yMin, zMin] # set header offset
lasOutput.header.scale = [0.001, 0.001, 0.001] #3 decimal point precision, can change as necessary
lasOutput.x = data1[:,0]
lasOutput.y = data1[:,1]
lasOutput.z = data1[:,2]

lasOutput.close()




# hdr = laspy.header.Header()

# lasOutput = laspy.file.File("lidarcar.las", mode = 'w', header = hdr)


# xMin = np.floor(np.min(data[:,0])) #use numpy to get int min values of x y z
# yMin = np.floor(np.min(data[:,1]))
# zMin = np.floor(np.min(data[:,2]))

# lasOutput.header.offset = [xMin, yMin, zMin] # set header offset
# lasOutput.header.scale = [0.001, 0.001, 0.001] #3 decimal point precision, can change as necessary
# lasOutput.x = data[:,0]
# lasOutput.y = data[:,1]
# lasOutput.z = data[:,2]

# lasOutput.close()
```
